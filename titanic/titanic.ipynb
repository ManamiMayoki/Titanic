```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_data=pd.read_csv('data/train.csv')
titanic_data.head()
titanic_data.info()
titanic_data.describe()
```

```python
titanic_data
```

```python
import seaborn as sns
```

```python
sns.heatmap(titanic_data.corr(numeric_only=True), cmap="YlGnBu")
plt.show()
```

```python
col=titanic_data['Age']
print("Mean:", col.mean())
print("Median:", col.median())
print("Standard Deviation:", col.std())
```

```python
import matplotlib.pyplot as plt

# Histogram for Age
plt.figure(figsize=(8, 4))
sns.histplot(titanic_data['Age'].dropna(), kde=True, color='teal')
plt.title('Distribution of Age')
plt.show()

# Boxplot for Fare (shows outliers)
plt.figure(figsize=(8, 4))
sns.boxplot(x=titanic_data['Fare'], color='salmon')
plt.title('Fare Boxplot')
plt.show()
```
